```python
#来自https://morvanzhou.github.io/tutorials/python-basic/threading/4-queue/
#如何添加线程以及线程运用
#当然所谓的多线程只是节省了读写数据的时间，可以使用多核运算进一步节省时间
import threading
import time
from queue import Queue #有关于多线程数值返回的包

def job(l,q):
    for i in range(len(l)):
        l[i] = l[i]**2
    #return l 在线程返回结果中是不行的
    q.put(l) #结果进入队列中

def multithreading():
    q = Queue() #这是一个先进先出FIFO队列， 而.LifoQueue为先进后出
    data = [[1,2,3],[2,3,4]]
    threads = []
    for i in range(2):
        t = threading.Thread(target = job, args = (data[i],q))
        #创建一个新的线程，arg为参数
        t.start()
        threads.append(t)#加入到threads这个列表中
    for thread in threads: #此时threads中有两个新的线程
        thread.join()
    results = []
    for _ in range(2):
        results.append(q.get())
    print(results)

def T1_job():
    print('This is the T1 thread, the thread name is %s \n'
          %threading.current_thread())
    print('T1_job starts \n')
    for i in range(10):
        time.sleep(0.1) #休眠0.1秒
    print('T1 finishs \n')

def T2_job():
    print('T2 starts.\n')
    print('T2 finishs.\n')

def job1():
    global A
    lock.acquire()
    for i in range(10):
        A += 1
        print('job1',A)
    lock.release()
        
def job2():
    global A
    lock.acquire()
    for i in range(10):
        A += 10
        print('job2',A)
    lock.release()
    
def main():
    print(threading.active_count()) #当前激活线程
    print(threading.enumerate()) #线程名字
    print(threading.currentThread()) #当前运行的线程
    print('\n')
    
    #创建新线程以及线程开始，结束
    T1 = threading.Thread(target = T1_job)
    T2 = threading.Thread(target = T2_job)
    T1.start()
    T2.start()
    T1.join()
    T2.join() #.join只对主线程有效，支线之间没有先后执行顺序，执行完此后才能执行接下来的语句
    print('All done. \n')
    
    #关于线程值返回
    multithreading()
    
    #线程lock,运行完一个线程之后再运行另一个
    
    t1 = threading.Thread(target = job1)
    t2 = threading.Thread(target = job2)
    t1.start()
    t2.start()
    
if __name__ == '__main__':
    A = 0
    lock = threading.Lock()
    main()
```
